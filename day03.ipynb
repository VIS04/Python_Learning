```python
Children = input("Enter number of children")
if Children == "0":
    Children = None
    print("You have no children, spend all your money and enjoy life")
else:
    Children = int(Children)
    if Children ==1:
        print("You have 1 child, be a responsible parent, Enjoy life with family!")
    else:
        print("You have ", Children, "children, be a responsible parent, Enjoy life with family!")
print(type(Children))
```
