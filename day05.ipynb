# Conditional Statements

```python
mark = int (input("Enter Mark"))
if mark == 100:
    print('go to ootty')
elif mark > 80:
    print('buy phone')
elif mark > 60:
    print ('buy cycle')
else:
    print ('Get Out')
```

```python
#Nested conditional statement
num = int(input('Enter a number'))
k = [1,2,3,4,5,0]

if num in k:
    if num == 0:
        print(f'Number {num} is in the list')
    elif num%2 ==0:
        print(f' Number {num} is even')
    else:
        print(f'Number {num} is odd')
else:
    print (f'Number is not in the list')
```

```python
number = int (input("Enter a number"))
if number%5 ==0 and number%3 ==0:
    print('fizzbuzz')
elif number%5 ==0:
    print ('fizz')
elif number%3 ==0:
    print ('buzz')
else:
    print('Number is not divisible by 5 or 3')
```

# Loop

```python
num = [1,2,3,4,5]
for i in num:
    print(i)
```

```python
num = ['apple','orange','mango']
for i in num:
    print(i)
```

```python
num = ['apple','orange','mango']
for i in num:
    if num.index(i) == 0:
        continue
    else:
        print(i)
```

```python
num = ['apple','orange','mango']
for i in num:
    if num.index(i) == 2:
        break
    else:
        print(i)
```

```python
num = ['apple','orange','mango']
for i in num:
    if num.index(i) == 2:
        pass
    else:
        print(i)
```
